```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

'''
Si depues se tiene que trabajar con mas tiendas, solo agregas la dupla de tienda y url del archivo y debe funcionar todo super chido'''
tiendas = {
    "tienda1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "tienda2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "tienda3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "tienda4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"}

def suma_columna_tienda(tienda, categoria): #Hace la suma de todas las filas de una sola columna(categoria)
    return round(tienda[categoria].sum(),2)

def promedio_calificacion(tienda):
    return round(tienda['Calificación'].mean(), 2) #Redondea y promedia las calificaciones de las tiendas

def venta_por_columna(tienda, columna): #Hace un contador que permite ver cuantas ventas por columna seleccionada hay
    venta = {}
    for index, row in tienda.iterrows():
        if row[columna] not in venta.keys():
            venta[row[columna]] = 0
        venta[row[columna]] += 1
    return venta

def ingreso_neto_tienda(data: dict, clave_minuendo: str, clave_sustraendo: str, clave_resultado: str = 'resultado') -> None:
    for key, subdict in data.items():
        minuendo = subdict.get(clave_minuendo, 0.0)
        sustraendo = subdict.get(clave_sustraendo, 0.0)
        subdict[clave_resultado] = round(minuendo - sustraendo, 2)

def comparar_ingresos_tiendas(tiendas):
    print("Comparación de Ingresos por Tienda:\n")
    for tienda, ruta in tiendas.items():
        df = pd.read_csv(ruta)
        precio_total = df['Precio'].sum()
        costo_envio_total = df['Costo de envío'].sum()
        ingreso_neto = precio_total - costo_envio_total
        print(f"{tienda} -> Precio total: ${precio_total:.2f}, "
              f"Envío total: ${costo_envio_total:.2f}, "
              f"Ingreso neto: ${ingreso_neto:.2f}")

def producto_mas_rentable(tienda):
    productos = {}
    for x, fila in tienda.iterrows():
        producto = fila['Producto']
        ingreso = fila['Precio']
        costo_envio = fila['Costo de envío']
        if producto not in productos:
            productos[producto] = {'ventas': 0, 'precio_total': 0.0, 'envio_total': 0.0}
        productos[producto]['ventas'] += 1
        productos[producto]['precio_total'] += ingreso
        productos[producto]['envio_total'] += costo_envio

    rendimientos = []
    for nombre, datos in productos.items():
        ingreso_neto = datos['precio_total'] - datos['envio_total']
        rendimiento = ingreso_neto / datos['ventas']
        rendimientos.append((nombre, round(rendimiento, 2)))

    def clave_orden(tupla): #Funcion que retorna el segundo elemento de una tupla, en este caso el rendimiento
        return tupla[1]

    rendimientos.sort(key=clave_orden, reverse=True)#ordena las tuplas segun el valor de su segundo elemento
    return rendimientos[0]

def cinco_productos_mas_rentables(tienda):
    productos = {}
    for x, fila in tienda.iterrows():
        producto = fila['Producto']
        ingreso = fila['Precio']
        costo_envio = fila['Costo de envío']
        if producto not in productos:
            productos[producto] = {'ventas': 0, 'precio_total': 0.0, 'envio_total': 0.0}
        productos[producto]['ventas'] += 1
        productos[producto]['precio_total'] += ingreso
        productos[producto]['envio_total'] += costo_envio

    rendimientos = []

    for nombre, datos in productos.items():
        ingreso_neto = datos['precio_total'] - datos['envio_total']
        rendimiento = ingreso_neto / datos['ventas']
        rendimientos.append((nombre, round(rendimiento, 2)))

    def clave_orden(tupla):
        return tupla[1]

    rendimientos.sort(key=clave_orden, reverse=True)
    return rendimientos[:5]

def producto_mas_costoso_de_mantener(tienda):
    productos = {}
    for x, fila in tienda.iterrows():
        producto = fila['Producto']
        ingreso = fila['Precio']
        costo_envio = fila['Costo de envío']
        if producto not in productos:
            productos[producto] = {'ventas': 0, 'precio_total': 0.0, 'envio_total': 0.0}
        productos[producto]['ventas'] += 1
        productos[producto]['precio_total'] += ingreso
        productos[producto]['envio_total'] += costo_envio

    rendimientos = []
    for nombre, datos in productos.items():
        ingreso_neto = datos['precio_total'] - datos['envio_total']
        rendimiento = ingreso_neto / datos['ventas']
        rendimientos.append((nombre, round(rendimiento, 2)))

    def clave_orden(tupla): #Funcion que retorna el segundo elemento de una tupla, en este caso el rendimiento
        return tupla[1]

    rendimientos.sort(key=clave_orden)#ordena las tuplas segun el valor de su segundo elemento
    return rendimientos[0]

def cinco_productos_mas_costosos_de_mantener(tienda):
    productos = {}
    for x, fila in tienda.iterrows():
        producto = fila['Producto']
        ingreso = fila['Precio']
        costo_envio = fila['Costo de envío']
        if producto not in productos:
            productos[producto] = {'ventas': 0, 'precio_total': 0.0, 'envio_total': 0.0}
        productos[producto]['ventas'] += 1
        productos[producto]['precio_total'] += ingreso
        productos[producto]['envio_total'] += costo_envio

    rendimientos = []

    for nombre, datos in productos.items():
        ingreso_neto = datos['precio_total'] - datos['envio_total']
        rendimiento = ingreso_neto / datos['ventas']
        rendimientos.append((nombre, round(rendimiento, 2)))

    def clave_orden(tupla):
        return tupla[1]

    rendimientos.sort(key=clave_orden)
    return rendimientos[:5]
##Siento que pude haberlo dejado más eficiente por que algunas de estas funciones hacen lo mismo, pero ya no se me ocurrió como y no encontré googleando
def suma_neta_cinco_menos_rentables(tienda):
    productos = {}
    for _, fila in tienda.iterrows():
        nombre = fila['Producto']
        ingreso = fila['Precio']
        costo_envio = fila['Costo de envío']

        if nombre not in productos:
            productos[nombre] = {'ventas': 0, 'precio_total': 0.0, 'envio_total': 0.0}
        productos[nombre]['ventas'] += 1
        productos[nombre]['precio_total'] += ingreso
        productos[nombre]['envio_total'] += costo_envio

    netos_por_producto = []
    for nombre, datos in productos.items():
        ingreso_neto_prod = datos['precio_total'] - datos['envio_total']
        netos_por_producto.append((nombre, round(ingreso_neto_prod, 2)))

    def clave_orden(tupla):
        return tupla[1]

    netos_por_producto.sort(key=clave_orden)

    cinco_menos_vendidos = netos_por_producto[:5]
    suma_cinco_menos = 0.0
    for x, ingreso_neto in cinco_menos_vendidos:
        suma_cinco_menos += ingreso_neto

    return round(suma_cinco_menos, 2)

def preparar_graph(data, x_col=None, y_col=None): #Prepara los datos ingresados (dict, list, tupla, dataframe) para hacer un gráfico
    etiquetas = []
    valores = []
    if isinstance(data, dict): #Si es un diccionario
        for clave, valor in data.items():
            etiquetas.append(clave)
            valores.append(valor)
    elif isinstance(data, list) and all(isinstance(i, (tuple, list)) and len(i) == 2 for i in data): #Si es una lista de tuplas o lista de listas
        for x, y in data:
            etiquetas.append(x)
            valores.append(y)
    elif hasattr(data, 'loc') and x_col and y_col: #Si es un DataFrame y se pasan columnas
        etiquetas = data[x_col].tolist()
        valores = data[y_col].tolist()
    else:
        raise ValueError(
            "Formato de datos no soportado. Usa dict, lista de pares o DataFrame con columnas especificadas.")
    return etiquetas, valores

def crear_pie_graph(data, labels, title=None, autopct='%1.1f%%', shadow=False, explode=None, startangle=0):#Crea un gráfico de pay
    plt.figure(figsize=(8, 8))#Crea la figura donde se pinta
    plt.pie(data, labels=labels, autopct=autopct, shadow=shadow, explode=explode, startangle=startangle)
    if title:
        plt.title(title)
    plt.axis('equal')
    plt.show()

def crear_grouped_bar_graph(etiquetas, lista_de_valores, nombres_grupo,
                            titulo="", xlabel="", ylabel=""):
    num_tiendas = len(etiquetas)
    num_grupos = len(lista_de_valores)

    x = np.arange(num_tiendas)
    ancho_barra = 0.8 / num_grupos  # Ajustar el ancho para que quepan todas

    fig, ax = plt.subplots(figsize=(12, 6))  # Ajustar tamaño del gráfico

    for i, valores in enumerate(lista_de_valores):
        desplazamiento = (i - num_grupos / 2) * ancho_barra + ancho_barra / 2
        posiciones = x + desplazamiento
        # Solo graficamos si hay algún valor significativo
        if any(valores):
            ax.bar(posiciones, valores, width=ancho_barra, label=nombres_grupo[i])

    ax.set_title(titulo)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_xticks(x)
    ax.set_xticklabels(etiquetas, rotation=45)
    ax.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), fontsize=8, ncol=1, title="Categorías")

    plt.tight_layout()
    plt.show()

def crear_bar_graph_simple(data: dict, title: str, xlabel: str = 'Tienda', ylabel: str = 'Valor'):
    etiquetas, valores = preparar_graph(data) #Prepara etiquetas y valores para graficar

    plt.figure(figsize=(10, 5)) #Crea la figura donde se va a dibujar el gráfico
    barras = plt.bar(etiquetas, valores)  #Dibuja las barras con las etiquetas y valores

    for barra in barras: #Añade etiquetas numéricas a cada barra (valor sobre cada barra)
        altura = barra.get_height()
        plt.annotate(f'{altura:.2f}', xy=(barra.get_x() + barra.get_width() / 2, altura),
                     xytext=(0, 3), textcoords="offset points",
                     ha='center', va='bottom')

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.tight_layout()
    plt.show()

def crear_grouped_bar_para_productos(resultados: dict, subkey: str, titulo: str = None,
                                       xlabel: str = "Tienda", ylabel: str = "Cantidad",
                                       top_n: int = None, menos_vendidos: bool = False):
    # Obtener lista ordenada de tiendas
    tiendas_ordenadas = sorted(resultados.keys())  # Ordena las tiendas alfabéticamente

    # Obtener conjunto de todas las categorías posibles
    conjunto_categorias = set()  # Agrupa todas las categorías encontradas en todas las tiendas (sin duplicados)
    for tienda in tiendas_ordenadas:
        resultado_dict = resultados[tienda].get(subkey, {})
        conjunto_categorias.update(resultado_dict.keys())

    # Si se quiere limitar a los productos más o menos vendidos
    if top_n is not None:
        # Calcular el total de cada categoría sumando en todas las tiendas
        conteo_total_por_categoria = {}
        for etiqueta in conjunto_categorias:
            total = 0
            for tienda in tiendas_ordenadas:
                total += resultados[tienda].get(subkey, {}).get(etiqueta, 0)
            conteo_total_por_categoria[etiqueta] = total

        # Ordenar según total de ventas (descendente o ascendente según lo que se desee)
        etiquetas_ordenadas = sorted(
            conteo_total_por_categoria,
            key=conteo_total_por_categoria.get,
            reverse=not menos_vendidos  # Si menos_vendidos=True, ordena ascendente
        )[:top_n]  # Selecciona los N primeros
    else:
        # Si no se especifica top_n, usar todas las etiquetas ordenadas alfabéticamente
        etiquetas_ordenadas = sorted(conjunto_categorias)

    # Construir lista de valores por categoría (una serie por categoría)
    lista_de_valores = []
    for etiqueta in etiquetas_ordenadas:
        valores_serie = []
        for tienda in tiendas_ordenadas:
            conteo = resultados[tienda].get(subkey, {}).get(etiqueta, 0)
            valores_serie.append(conteo)
        lista_de_valores.append(valores_serie)

    # Generar título si no se proporcionó uno
    if not titulo:
        titulo = f"Comparación de {subkey}"

    # Llamar a la función de graficado
    crear_grouped_bar_graph(etiquetas=tiendas_ordenadas,
                            lista_de_valores=lista_de_valores,
                            nombres_grupo=etiquetas_ordenadas,
                            titulo=titulo,
                            xlabel=xlabel,
                            ylabel=ylabel)

def construir_ingreso_neto(tiendas: dict, subkey: str = 'ingreso_neto') -> dict:
    resultados = {}
    for nombre_tienda, ruta_csv in tiendas.items():
        df = pd.read_csv(ruta_csv)
        ingreso_neto = df['Precio'].sum() - df['Costo de envío'].sum()
        resultados[nombre_tienda] = {subkey: ingreso_neto}
    return resultados

def mostrar_top5_bot5_productos_tienda(tienda_nombre: str, conteos: dict):
    if not conteos:
        print(f"{tienda_nombre}: No hay datos de venta por producto.\n")
        return

    lista_productos = list(conteos.items())
    lista_ordenada_desc = sorted(lista_productos, key=lambda x: x[1], reverse=True)
    lista_ordenada_asc = sorted(lista_productos, key=lambda x: x[1])

    top5 = lista_ordenada_desc[:5]
    bot5 = lista_ordenada_asc[:5]
    print("El top 5 de productos más vendidos:")
    for producto, cantidad in top5:
        print(f"  • {producto}: {cantidad} unidades")

    print("\nEl top 5 de productos menos vendidos:")
    for producto, cantidad in bot5:
        print(f"  • {producto}: {cantidad} unidades")

########################################################################################################################
def main():
    resultados = {}
    for key, value in tiendas.items():
        tienda = pd.read_csv(value)
        resultados[key] = {}
        resultados[key]['ingreso_total'] = suma_columna_tienda(tienda, 'Precio')
        resultados[key]['venta_por_categoria'] = venta_por_columna(tienda, 'Categoría del Producto')
        resultados[key]['venta_por_producto'] = venta_por_columna(tienda, 'Producto')
        resultados[key]['promedio_calificacion'] = promedio_calificacion(tienda)
        resultados[key]['envio_tienda'] = suma_columna_tienda(tienda, 'Costo de envío')

    ingreso_neto_tienda(resultados, 'ingreso_total', 'envio_tienda', 'ingreso_neto')
    for key, value in resultados.items():
        tienda = pd.read_csv(tiendas[key])
        suma_peores5 = suma_neta_cinco_menos_rentables(tienda)
        ingreso_neto_filtrado = round(value['ingreso_neto'] - suma_peores5, 2)

        print(f'---Estadísticas de {key}---\n')
        # Ingresos y egresos
        print(f'Ingreso por ventas:                        ${value["ingreso_total"]:,.2f}')
        print(f'Egreso por envíos:                         ${value["envio_tienda"]:,.2f}')
        print(f'Ingreso neto original:                     ${value["ingreso_neto"]:,.2f}')
        print(f'Suma neta de 5 productos menos rentables:  ${suma_peores5:,.2f}')
        print(f'Ingreso neto filtrado (sin esos 5):        ${ingreso_neto_filtrado:,.2f}\n')

        # Calificación promedio
        print(f'Calificación promedio de productos:        {value["promedio_calificacion"]:.2f}\n')

        # Ventas por categoría
        print('Ventas por categoría:')
        print(f'  {value["venta_por_categoria"]}')
        print(f'  (Ver gráfica: {key}_venta_por_categoria)\n')
        etiquetas, valores = preparar_graph(value['venta_por_categoria'])
        crear_pie_graph(valores, etiquetas, title=f'{key}_venta_por_categoria')

        # Ventas por producto
        print('\nVentas por producto:')
        print(f'  {value["venta_por_producto"]}')
        print(f'  (Ver gráfica: {key}_venta_por_producto)\n')
        etiquetas, valores = preparar_graph(value['venta_por_producto'])
        crear_pie_graph(valores, etiquetas, title=f'{key}_venta_por_producto', shadow=True)
        mostrar_top5_bot5_productos_tienda(key, value['venta_por_producto'])
        print()

        # Productos más y menos rentables
        mejor_producto, rendimiento_positivo = producto_mas_rentable(tienda)
        print(f'Producto más rentable en {key}:')
        print(f'  • {mejor_producto} → Rendimiento por unidad: ${rendimiento_positivo:,.2f}\n')

        print('5 productos más rentables:')
        for prod, rend in cinco_productos_mas_rentables(tienda):
            print(f'  - {prod}: ${rend:,.2f} por unidad')
        print()

        peor_producto, rendimiento_negativo = producto_mas_costoso_de_mantener(tienda)
        print(f'Producto más costoso de mantener en {key}:')
        print(f'  • {peor_producto} → Rendimiento por unidad: ${rendimiento_negativo:,.2f}\n')

        print('5 productos menos rentables en promedio:')
        for producto, rendimiento in cinco_productos_mas_costosos_de_mantener(tienda):
            print(f'  - {producto}: ${rendimiento:,.2f} por unidad')

        print('\n' + '-' * 60 + '\n')  # Imprime 60 guiones

    # Gráficos comparativos y comparación de ingresos entre tiendas
    resultados_neto = construir_ingreso_neto(tiendas)
    crear_bar_graph_simple(
        data={k: v['ingreso_neto'] for k, v in resultados_neto.items()},
        title='Ingreso Neto por Tienda',
        xlabel='Tienda',
        ylabel='Ingreso Neto'
    )
    crear_grouped_bar_para_productos(
        resultados, subkey='venta_por_producto', top_n=5, menos_vendidos=False,
        titulo='Más ventas', xlabel='Tienda', ylabel='Cantidad Vendida'
    )
    crear_grouped_bar_para_productos(
        resultados, subkey='venta_por_producto', top_n=5, menos_vendidos=True,
        titulo='Menos ventas', xlabel='Tienda', ylabel='Cantidad Vendida'
    )
    comparar_ingresos_tiendas(tiendas)

    # Determinar tienda menos rentable
    menor_tienda = None
    menor_neto = None
    for key, value in resultados.items():
        neto = value['ingreso_neto']
        if menor_neto is None or neto < menor_neto:
            menor_neto = neto
            menor_tienda = key
    print(f'La tienda menos rentable es {menor_tienda} con ingreso neto de: ${menor_neto:,.2f}\n')
    print()

    # Observaciones y recomendaciones
    print('Observaciones y recomendaciones:\n')
    print('1. La rentabilidad neta por tienda es:')
    for tienda, valores in resultados.items():
        print(f'   - {tienda}: ${valores["ingreso_neto"]:,.2f}')
    print(f'\n2. Conviene enfocar esfuerzos en productos con un alto margen de ganancia unitario (TV LED UHD 4K, Refrigerador, iPhone 15).')
    print('3. No priorizar ventas en los 5 productos menos rentables debido a su bajo aporte neto.')
    print('4. Hay que mejorar la calidad de las reseñas, principalmente para la tienda1, que aun siendo la que tiene mayor cantidad de ventas, en promedio tiene las reseñas más bajas .')
    print('5. Analizar constantemente que productos cuesta más trabajo mantener y se venden menos.')
    print('6. Controlar inventario de productos de bajo margen de ganancia para evitar costos de almacenamiento.\n')
    print('Estadísticamente  habría que vender la tienda 4 para iniciar un nuevo emprendimiento.\nPero este análisis nos permitió ver en que podemos mejorar las otras tiendas')


if __name__ == "__main__":
    main()
```
